# **🏠House Price Prediction with Linear Regression**

This project contains data preparation, model training & perfomance evaluation in order to predict house prices and effectiveness in minimizing differences between variables.

Our first step must be to prepare the dataset. This process includes eliminating duplicated or null info and correcting misspelled words if is necessary.

```python
# Importing & viewing the dataset:
import pandas as pd
df_house = pd.read_csv("home_dataset.csv" , encoding="utf-8-sig")
df_house.info
```

```python
# Dropping the value with the "1" price:
df_house = df_house.drop(df_house.index[22])
```

# **⏭ Model explanation:**
Linear regression is used to show the relationship between an independent variable and a dependent variable. It assumes a linear relationship, meaning the output changes at a constant rate as the input changes. This relationship is represented by a straight line, often referred to as the best-fit line.

The reason behind using this model is because of the ability to provide clear insights and easiness for implementation. For house price prediction, this model is perfect for revealing how the value of the property and its size are related to show the actual price of the house.

```python
# Importing libraries & preparing the info for the model
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
data = df_house
house_sizes = data ['HouseSize'].values
house_prices = data['HousePrice'].values
```

```python
# Training the model
x_train, x_test, y_train, y_test = train_test_split(house_sizes, house_prices, test_size=0.2, random_state=42)

x_train = x_train.reshape(-1, 1)
x_test = x_test.reshape(-1, 1)

model = LinearRegression()
model.fit(x_train, y_train)
```

# **📈Chart intepretation:**

- 50% of the values almost match with the prediction line.

- The other half are simply out the line.

- The price of the house will be proportional to its size. This means that the bigger the house, the more money it will cost.



```python
# Visualizing the model:
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="darkgrid")


fig, ax = plt.subplots(1, 2, figsize=(15, 6))

# Graphic #1
sns.scatterplot(x=house_sizes, y=house_prices, marker='o', color='blue', ax=ax[0])
ax[0].set_title('House Prices vs. House Size', fontsize=12, fontweight='bold')
ax[0].set_xlabel('House Size (sq.ft)')
ax[0].set_ylabel('House Price ($)')


predictions = model.predict(x_test)

# Graphic #2
ax[1].scatter(x_test, y_test, marker='o', color='blue', label='Actual Prices')
ax[1].plot(x_test, predictions, color='red', linewidth=2, label='Predicted Prices')
ax[1].set_title('Property Price Prediction\nwith Linear Regression', fontsize=12, fontweight='bold')
ax[1].set_xlabel('House Size (sq.ft)')
ax[1].set_ylabel('House Price (millions $)')
ax[1].legend()


plt.tight_layout()
plt.show()
```

# **💡Key Insights:**

- Linear Regression provides a simple yet powerful baseline model for regression in house pricing problems.

- It offers strong interpretability, allowing us to understand how each feature influences price.

- Despite its simplicity, it is effective for datasets where relationships between variables are approximately linear.
